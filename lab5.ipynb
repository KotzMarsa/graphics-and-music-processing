```python
%matplotlib inline
from scipy import misc
from scipy import ndimage
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
lena = misc.imread('images/LENA_512.jpg')
```

```python
def change_value(value):
    if value < 0:
        return 0
    elif value > 255:
        return 255
    else:
        return value
```

```python
def greyscale(picture):
    grey_picture = np.copy(picture)
    for row in grey_picture:
        for col in row:
            col[0] = col[1] = col[2] = (int(col[0]) + int(col[1]) + int(col[2]))/3
    return grey_picture
```

```python
def uniform_noise(picture, level=100, probability=0.3):
    noised_picture = np.copy(picture)
    for row in noised_picture:
        for col in row:
            if(random.random() <= probability):
                noise_value = int(np.random.uniform(-level,level))
                col[0] = change_value(col[0] + noise_value)
                col[1] = change_value(col[1] + noise_value)
                col[2] = change_value(col[2] + noise_value)
    return noised_picture
```

```python
uniform_noise_lena = uniform_noise(lena)
plt.imshow(uniform_noise_lena)
plt.show()
```

```python
uniform_noise_grey_lena = uniform_noise(greyscale(lena))
plt.imshow(uniform_noise_grey_lena)
plt.show()
```

```python
def normal_noise(picture, m=0, s=100, probability=0.3):
    noised_picture = np.copy(picture)
    for row in noised_picture:
        for col in row:
            if(random.random() <= probability):
                noise_value = int(np.random.normal(m,s))
                col[0] = change_value(col[0] + noise_value)
                col[1] = change_value(col[0] + noise_value)
                col[2] = change_value(col[0] + noise_value)
    return noised_picture
```

```python
normal_noise_lena = normal_noise(lena)
plt.imshow(normal_noise_lena)
plt.show()
```

```python
normal_noise_grey_lena = normal_noise(greyscale(lena))
plt.imshow(normal_noise_grey_lena)
plt.show()
```

```python
def salt_and_paper_noise(picture, probability=0.1):
    noised_picture = np.copy(picture)
    for row in noised_picture:
        for col in row:
            if(random.random() <= probability):
                col[0] = col[1] = col[2] = random.choice([0, 255])
    return noised_picture
```

```python
salt_and_paper_noise_lena = salt_and_paper_noise(lena)
plt.imshow(salt_and_paper_noise_lena)
plt.show()
```

```python
salt_and_paper_noise_noise_grey_lena = salt_and_paper_noise(greyscale(lena))
plt.imshow(salt_and_paper_noise_noise_grey_lena)
plt.show()
```
