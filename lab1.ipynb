```python
%matplotlib inline
from scipy import misc
import matplotlib.pyplot as plt
import numpy as np
```

```python
lena = misc.imread('images/LENA_512.jpg')
```

```python
type(lena)
```

```python
lena.shape, lena.dtype
```

```python
plt.imshow(lena)
plt.show()
```

```python
lena_red   = np.copy(lena)
lena_green = np.copy(lena)
lena_blue  = np.copy(lena)
```

```python
for row in lena_red:
    for col in row:
        col[1] = 0
        col[2] = 0
```

```python
plt.imshow(lena_red)
plt.show()
```

```python
for row in lena_green:
    for col in row:
        col[0] = 0
        col[2] = 0
```

```python
plt.imshow(lena_green)
plt.show()
```

```python
for row in lena_blue:
    for col in row:
        col[0] = 0
        col[1] = 0
```

```python
plt.imshow(lena_blue)
plt.show()
```
