```python
%matplotlib inline
from scipy import misc
import matplotlib.pyplot as plt
import matplotlib.colors
import numpy as np
```

```python
lena = misc.imread('images/LENA_512.jpg')
ship = misc.imread('images/statek_640_505.jpg')
mlecz = misc.imread('images/mlecz.gif')
medl = misc.imread('images/meduza_l.jpg')
medr = misc.imread('images/meduza_r.jpg')
japl = misc.imread('images/japan_l.jpg')
japr = misc.imread('images/japan_r.jpg')
```

```python
def zeruj (picture, a, b):
    newpicture = np.copy(picture)
    for x, row in enumerate(newpicture):
        for y, col in enumerate(row):
            newpicture[x][y][a] = 0
            newpicture[x][y][b] = 0
    return newpicture
```

```python
s1 = zeruj(mlecz,0,1)
plt.imshow(s1)
plt.show()
s2 = zeruj(mlecz,1,2)
plt.imshow(s2)
plt.show()
s3 = zeruj(mlecz,2,0)
plt.imshow(s3)
plt.show()
```

```python
def lewe (picture):
    newpicture = np.copy(picture)
    for x, row in enumerate(newpicture):
        for y, col in enumerate(row):
            newpicture[x][y][0] = 0.299*picture[x][y][0] + 0.587*picture[x][y][1] + 0.114*picture[x][y][2]
            newpicture[x][y][1] = 0
            newpicture[x][y][2] = 0
    return newpicture
```

```python
def prawe (picture):
    newpicture = np.copy(picture)
    for x, row in enumerate(newpicture):
        for y, col in enumerate(row):
            newpicture[x][y][0] = 0
            newpicture[x][y][1] = 0.299*picture[x][y][0] + 0.587*picture[x][y][1] + 0.114*picture[x][y][2]
            newpicture[x][y][2] = 0.299*picture[x][y][0] + 0.587*picture[x][y][1] + 0.114*picture[x][y][2]
    return newpicture
```

```python
s1 = lewe(mlecz)
plt.imshow(s1)
plt.show()
s1 = prawe(mlecz)
plt.imshow(s1)
plt.show()
```
