```python
%matplotlib inline
from scipy import misc
import matplotlib.pyplot as plt
import matplotlib.colors
import numpy as np
```

```python
lena = misc.imread('images/LENA_512.jpg')
lake = misc.imread('images/lake.jpg')
eagle = misc.imread('images/eagle.jpg')
boldt = misc.imread('images/boldt.jpg')
rain = misc.imread('images/rain.jpg')
tex = misc.imread('images/tex.jpg')
kula = misc.imread('images/kula.jpg')
s1 = misc.imread('images/s1.png')
s2 = misc.imread('images/s2.png')
krol1 = misc.imread('images/krol1.png')
krol2 = misc.imread('images/krol2.png')
```

```python
def EL(picture,z):
    saturation_picture = np.copy(picture)
    matplotlib.colors.rgb_to_hsv(saturation_picture)
    for row in saturation_picture:
        for col in row:
            col[0] = min((col[0] + z),255)
    matplotlib.colors.hsv_to_rgb(saturation_picture)
    return saturation_picture
```

```python
lena_sat = EL(lena, 100)
plt.imshow(lena_sat)
plt.show()
```

```python
def negative(picture):
    negative_picture = np.copy(picture)
    for row in negative_picture:
        for col in row:
            col[0] = 255 - col[0]
            col[1] = 255 - col[1]
            col[2] = 255 - col[2]
    return negative_picture
```

```python
def znajdzroznice(pic1, pic2):
    new_picture = np.full(np.shape(pic1), 255, dtype=np.uint8)
    for x, row in enumerate(pic2):
        for y, col in enumerate(row):
            if (col[0] != pic1[x][y][0] or col[1] != pic1[x][y][1] or col[2] != pic1[x][y][2]):
                new_picture[x][y][0] = pic2[x][y][0]
                new_picture[x][y][1] = pic2[x][y][1] 
                new_picture[x][y][2] = pic2[x][y][2] 
    return new_picture
    
```

```python
s3 = znajdzroznice(s1,s2)
plt.imshow(s3)
plt.show()
```

```python
s3 = znajdzroznice(krol2,krol1)
plt.imshow(s3)
plt.show()
```

```python
def jeden(pic1, pic2, w):
    new_picture = np.full(np.shape(pic1), 255, dtype=np.uint8)
    for x, row in enumerate(pic2):
        for y, col in enumerate(row):
            new_picture[x][y][0] = w*pic1[x][y][0] + (1-w)*pic2[x][y][0]
            new_picture[x][y][1] = w*pic1[x][y][1] + (1-w)*pic2[x][y][1]
            new_picture[x][y][2] = w*pic1[x][y][2] + (1-w)*pic2[x][y][2]
    return new_picture
        
```

```python
plt.imshow(lena, lake,10)
plt.show()
s3 = jeden(lena, lake, 10)
plt.imshow(s3)
plt.show()
```
